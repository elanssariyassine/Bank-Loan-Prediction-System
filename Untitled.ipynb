# Data Cleaning

```python
#Separé les variables categoriques et les variables numériques
cat_data=[]
num_data=[]

for i,c in enumerate(df.dtypes):
  if c==object:
    cat_data.append(df.iloc[:,i])
  else :
    num_data.append(df.iloc[:,i])
```

```python
num_data
```

```python
cat_data=pd.DataFrame(cat_data).transpose()
```

```python
cat_data
```

```python
num_data=pd.DataFrame(num_data).transpose()
```

```python
num_data
```

```python
#pour les valeurs categoriques on va remplassez les valeures null par les valeurs qui se repetent les plus 
cat_data=cat_data.apply(lambda x:x.fillna(x.value_counts().index[0]))
cat_data.isnull().sum().any()
```

```python
cat_data['Married'].value_counts()
```

```python
#pour les variables numeriques on va remplacer les valeures manquantes par la valeur précedente de la méme colonne
num_data.fillna(method='bfill',inplace=True)
num_data.isnull().sum().any()
```

```python
# transformer la colonne target from Y/N to 1/0
target_value={'Y':1,'N':0}
target=cat_data['Loan_Status']
cat_data.drop('Loan_Status',axis=1,inplace=True)
target=target.map(target_value)
target
```

```python
#remplasscer les valeurs categoriques par des valeures numeriques
le=LabelEncoder()
for i in cat_data:
    cat_data[i]=le.fit_transform(cat_data[i])
cat_data
```

```python
#supprimer loan-id
cat_data.drop('Loan_ID',axis=1,inplace=True)
```

```python
cat_data
```

```python
#concatener cat_data et num_data
X=pd.concat([cat_data,num_data],axis=1)
Y=target
```

# création du modéle

```python
#Diviser le dataset en train et test
sss=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train,test in sss.split(X,Y):
    X_train,X_test=X.iloc[train],X.iloc[test]
    Y_train,Y_test=Y.iloc[train],Y.iloc[test]
print('X_train taille:', X_train.shape)    
print('X_test taille:', X_test.shape) 
print('Y_train taille:', Y_train.shape) 
print('Y_test taille:', Y_test.shape) 
```

```python
# On va appliqu
models={
    'LogisticRegression':LogisticRegression(random_state=42),
    'KNeighborsClassifier':KNeighborsClassifier(),
    'DecisionTreeClassifier':DecisionTreeClassifier(max_depth=1,random_state=42),
    'GradientBoostingClassifier':GradientBoostingClassifier(n_estimators=33, learning_rate=0.1, max_depth=3, random_state=42)
}
#la fonction de précision
def accu(y_true,y_pred,retu=False):
   acc=accuracy_score(y_true,y_pred)
   if retu:
     return acc
   else:
        print(f'la precision du modéle est:{acc}')
# la fonction d'application des modeles
def train_test_eval(models,X_train,Y_train,X_test,Y_test):
        for name,model in models.items():
            print(name,':')
            model.fit(X_train,Y_train)
            accu(Y_test,model.predict(X_test))
            print('-'*30)
            
train_test_eval(models,X_train,Y_train,X_test,Y_test)        
```

```python
X_2=X[['Credit_History','Married','CoapplicantIncome','ApplicantIncome','Education','Self_Employed','LoanAmount']]
```

```python
#Diviser le dataset en train et test
sss=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train,test in sss.split(X_2,Y):
    X_train,X_test=X_2.iloc[train],X_2.iloc[test]
    Y_train,Y_test=Y.iloc[train],Y.iloc[test]
print('X_train taille:', X_train.shape)    
print('X_test taille:', X_test.shape) 
print('Y_train taille:', Y_train.shape) 
print('Y_test taille:', Y_test.shape)  
```

```python
train_test_eval(models,X_train,Y_train,X_test,Y_test)   
```
